```python
from sklearn.datasets import make_blobs
```

```python
X, y_true = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X[:,0], X[:,1], c=y_true)
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head(1)
```

```python
X = df[ [ "Annual Income (k$)", "Spending Score (1-100)" ] ].values
```

```python
from sklearn.cluster import KMeans
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
model = KMeans(n_clusters=3, random_state=42)
model.fit(X)
```

```python
y = model.predict(X)
```

```python
model.cluster_centers_
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X[:,0], X[:,1], c=y)
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], c='red', marker='X',s=250)
plt.show()
```

```python
# Parametre belirleme strateji
```

```python
# inertia -> Her elemann kendi kme merkezine olan uzaklklarnn karelerinin toplam.

model.inertia_
```

```python
# Elbow Method - Dirsek 
def elbow_method():
    inertias = []
    for k in range(2,11):
        model = KMeans(n_clusters=k, random_state=42)
        model.fit(X)
        inertias.append( model.inertia_ )
    plt.figure(figsize=(5,3))
    plt.plot(range(2,11), inertias)
    print(inertias)
```

```python
elbow_method()
```

```python
model2 = KMeans(n_clusters=5,random_state=42)
model2.fit(X)
y = model2.predict(X)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X[:,0], X[:,1], c=y)
plt.scatter(model2.cluster_centers_[:,0], model2.cluster_centers_[:,1], c='red', marker='X',s=250)
plt.show()
```
