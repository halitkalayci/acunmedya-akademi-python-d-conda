## Multiple Line Graph

```python
import numpy as np
x = np.linspace(0,10,100) #Linear Space
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(8,6)) #Figure -> Boş sayfa 8x6 inch
plt.plot(x,y1,label="sin(x)", color="red")
plt.plot(x,y2,label="cos(x)", color="green")
plt.legend() # -> bilgi alanı
plt.xlabel("Değerler")
plt.ylabel("Hesaplar (sin,cos)")
```

```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1) # 2 satır, 1 sutun, 1. grafik
plt.plot(x,y1)
plt.title("Sin(x)")

plt.subplot(2,1,2) # 2 satır, 1 sütun, 2. grafik
plt.plot(x,y2)
plt.title("Cos(x)")

plt.tight_layout() #Grafiklerin üst üste binmesini önler.
```

### Bar Grafiği


```python
categories = ["A","B","C"]
values = [23, 45, 12]
```

```python
plt.bar(categories, values, color=["red","green","blue","gray"])
plt.title("Kategorik Satış Grafiği")
plt.xlabel("Kategori")
plt.ylabel("Satış")
plt.show()
```

# Scatter Plot (Serpme Grafik)

```python
x = np.random.rand(50)
y = 2 * x + np.random.normal(0,0.1,50)
```

```python
plt.scatter(x,y, alpha=0.6)

# Line of Best Fit - Müthiş Çizgi
m, b = np.polyfit(x,y,1)

plt.plot(x, m*x+b, color="red")

plt.xlabel("Tahmin değeri")
plt.ylabel("Gerçek Değer")
```

# Histogram (Veri Dağılımı)

```python
data = np.random.randint(0,100,1000)
```

```python
plt.hist(data, bins=50, edgecolor="black",color="skyblue")
plt.xlabel("Değer")
plt.ylabel("Frekans")
plt.show()
## 100/30 => 3.33
## 100/10 => 10
```
