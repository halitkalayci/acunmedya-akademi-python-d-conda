```python
import matplotlib.pyplot as plt
```

```python
x = [ 1, 2, 3, 4, 5 ]
y = [ 500, 600, 300, 400, 100 ]
```

```python
plt.plot(x,y, color="g", marker="8", linestyle="dotted")
plt.grid()
plt.show() # Jupyter dışında bunu yazmanız gerekli.
#Çizgi Grafiği
```

## Multiple Line Graph

```python
import numpy as np
x = np.linspace(0,10,100) #Linear Space
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(8,6)) #Figure -> Boş sayfa 8x6 inch
plt.plot(x,y1,label="sin(x)", color="red")
plt.plot(x,y2,label="cos(x)", color="green")
plt.legend() # -> bilgi alanı
plt.xlabel("Değerler")
plt.ylabel("Hesaplar (sin,cos)")
```

```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1) # 2 satır, 1 sutun, 1. grafik
plt.plot(x,y1)
plt.title("Sin(x)")

plt.subplot(2,1,2) # 2 satır, 1 sütun, 2. grafik
plt.plot(x,y2)
plt.title("Cos(x)")
```
