Veri görselleştirme kütüphanesidir. Temel olarak 2D grafikler çizmek için kullanılır.

```python
import matplotlib.pyplot as plt
```

```python
x = [ 1, 2, 3, 4, 5 ]
y = [ 500, 600, 300, 400, 100 ]
```

```python
plt.plot(x,y, color="g", marker="8", linestyle="dotted")
plt.grid()
plt.show() # Jupyter dışında bunu yazmanız gerekli.
#Çizgi Grafiği
```

## Multiple Line Graph

```python
import numpy as np
x = np.linspace(0,10,100) #Linear Space
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(8,6)) #Figure -> Boş sayfa 8x6 inch
plt.plot(x,y1)
plt.plot(x,y2)
```
